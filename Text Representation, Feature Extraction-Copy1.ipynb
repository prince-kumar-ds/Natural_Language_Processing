```python
import numpy as np
import pandas as pd                                                                                                                                                                                                                                                              
```

```python
df=pd.DataFrame({'Text. No.':['D1','D2','D3','D4'],'Text':['people watch movie','movie watch movie','people write comment','movie write comment']})
```

```python
df
```

```python
from sklearn.preprocessing import MultiLabelBinarizer



df['Text Tokenized'] = df['Text'].apply(lambda x: x.split())

# Initialize the MultiLabelBinarizer for one-hot encoding
mlb = MultiLabelBinarizer()

# Perform one-hot encoding on the tokenized words
one_hot_encoded = mlb.fit_transform(df['Text Tokenized'])

# Create a new DataFrame with the one-hot encoded columns
one_hot_df = pd.DataFrame(one_hot_encoded, columns=mlb.classes_)

# Combine the original dataframe with the one-hot encoded columns
df_final = pd.concat([df, one_hot_df], axis=1)

# Display the result
df_final
```

# Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
bow=cv.fit_transform(df['Text'])
```

```python
# vocab
print(cv.vocabulary_)
```

```python
print(bow[0].toarray()) # Document 1
print(bow[1].toarray()) # Document 2
print(bow[2].toarray()) # Document 3
print(bow[3].toarray()) # Documnet 4
```

```python
# for new sentence
cv.transform(['movie watch and write comment of movie']).toarray() # prediction
```

```python
# Binary Bag of Words
cv=CountVectorizer(    binary=True,
)
bow=cv.fit_transform(df['Text'])
```

```python
print(bow[0].toarray()) # Document 1
print(bow[1].toarray()) # Document 2
print(bow[2].toarray()) # Document 3
print(bow[3].toarray()) # Documnet 4
```

# N-grams

```python
# Bi-grams
cv=CountVectorizer(ngram_range=(2,2))
bow=cv.fit_transform(df['Text'])
```

```python
# vocab
print(cv.vocabulary_)
```

```python
# Tri-grams
cv=CountVectorizer(ngram_range=(3,3))
bow=cv.fit_transform(df['Text'])
print(cv.vocabulary_)
```
